```python
import pandas as pd 
from bs4 import BeautifulSoup 
import requests
pd.set_option('display.max_columns',None)
import warnings
warnings.filterwarnings('ignore')
```

```python
# 1 the link
link= "https://www.billboard.com/charts/hot-100/"
```

```python
response=requests.get(link)
response.status_code #200 means Ok
```

```python
html=response.text
soup = BeautifulSoup(html, "html.parser")
```

```python
soup.find_all('li',class_='o-chart-results-list__item')
```

```python
songs= soup.find_all('li',class_='o-chart-results-list__item')
```

```python
data_songs={}
```

```python
for song in songs:
    title = song.find('h3', class_='c-title')
    artist = song.find('span', class_='c-label')
    if title and artist:
        data_songs[title.get_text(strip=True)] = artist.get_text(strip=True) #strip = true permet de supprimer les espacements
```

```python
df = pd.DataFrame(data_songs.items(), columns=['Title', 'Artist'])
```
